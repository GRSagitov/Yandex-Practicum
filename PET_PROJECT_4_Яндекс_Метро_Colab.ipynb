**Постановка задачи**

Команда Яндекс Метро провела рефакторинг мобильного приложения на Android — внесла правки в код. Чтобы выпустить новую версию, предварительно нужно:
* протестировать те части продукта, которых коснулись изменения;
* провести регрессионное тестирование и убедиться, что новую версию можно заливать в стор.

По результатам проверок необходимо предоставить отчёт о тестировании.

**Ссылки для работы:**
* [текущая версия приложения](https://code.s3.yandex.net/qa/files/yandexmetro-android-v2.13.apk), которую пользователи скачивают из стора;
* [готовящаяся сборка](https://code.s3.yandex.net/qa/files/yandexmetro-android-v3.6.apk);
* [требования к Яндекс Метро](https://code.s3.yandex.net/qa/files/Yandex_metro.pdf).

**Часть 1. Подготовка к функциональному тестированию**

Требования, которые затронул рефакторинг приложения, выделили полужирным шрифтом. Теперь нужно написать к ним тесты. Оформляю проверки в виде чек-листа.

**Часть 2. Подготовка к регрессионному тестированию**

Кроме проверок функциональности, затронутой рефакторингом, нужно провести регрессионное тестирование. Для этого напишу чек-лист, который учитывает особенности мобильного приложения:
1. Определяем, какая функциональность Яндекс Метро взаимодействует с мобильным устройством.
2. Создаю мобильные проверки, которые связаны с этой функциональностью.
3. Учтываю проверки, которые необходимы для любого мобильного приложения, — например, тестирование обновления.

**Пример проверок особенностей мобильного приложения:**

* поведение при переключении между типами сетей (3G, LTE, Wi-Fi);
* работа после восстановления связи;
* определение геолокации;
* работа при включении авиарежима;
* переключение между приложениями;
* поведение при входящем звонке/уведомлении;
* смена ориентации экрана;
* масштабирование интерфейса;
* включение режима энергосбережения;
* работа при низком заряде аккумулятора и т. д.

**Часть 3. Выполнение тестирования**

1. Тестирую мобильное приложение по своим чек-листам.
2. Завежу баг-репорты на последней вкладке своей копии Google-таблицы.

По итогам прошлого релиза команда определила, на каком устройстве возникает больше всего багов. Поэтому сейчас решили тестировать именно на этой конфигурации:
* Honor 8, ОС Android 9.0 Pie;
* разрешение экрана 1080×1920;
* диагональ 5.5.

**Важно:** Тестирование проводится на эмуляторе Android Studio.
