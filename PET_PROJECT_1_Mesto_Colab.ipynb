**Проект: Mesto - это интерактивная веб-страница, куда можно добавлять фотографии. Также фотографии можно оценивать лайком и удалять.**

**Задача:**
1. Провести регрессионное тестирование по готовым тест-кейсам.
2. Перепроверить баги, то есть провести ретест.

Инструменты: Тестовый стенд веб-страницы Mesto, Google Таблица в качестве TMS и BTS, Google Документ в качестве Отчета.

**Критерии сдачи. **
Ревьюер оценит итоговый проект по нескольким критериям:
1. Все ли проверки из тестового набора выполнены — для первой части;
2. Какие баги заведены — для первой части;
3. Верно ли проставлены статусы багов — для второй части.

**Часть 1. Регрессоинное тестирование.**

**Задачи:**
1. Познакомиться с сервисом Mesto.
2. Протестировать сервис по тест-кейсам и проставить статусы выполнения проверок.
3. Составить баг-репорты.

**Требования:**
https://praktikum.notion.site/Mesto-9f2cfaa209734d1f8cfa0c0db3d3049f

**Как выполнить:**
1. Открой таблицу со списком тест-кейсов для регрессионного тестирования и создай копию. https://docs.google.com/spreadsheets/d/1lomjKLJiPzBosP7bd4aQL5nb9XOQcbkhGn2M0Hsx1Dw/edit?gid=220888493#gid=220888493
2. Изучи тест-кейсы.
3. Зайди на тестовый стенд приложения Mesto и выполни тестирование по указанным тест-кейсам. https://code.s3.yandex.net/qa/files/mesto/index.html
4. В процессе тестирования отмечай результаты пройденных тестов: выбирай в колонке «Статус» PASSED, FAILED или другие подходящие статусы.
5. Если обнаружишь баг, запиши его в Google Таблицу. Вот шаблон — сделай копию и работай в ней. Обрати особое внимание на оформление баг-репортов:
  * чёткий заголовок, который кратко описывает суть проблемы (например, «Кнопка „Оплатить“ не работает в Chrome»);
  * корректный приоритет — укажи, насколько критична ошибка для пользователей;
  * точные шаги воспроизведения, чтобы разработчик мог повторить сценарий и увидеть ошибку;
  * ожидаемый результат vs фактический результат — объясни, что должно происходить и что происходит на самом деле;
  * вложения — прикрепи скриншоты или видео, если это поможет лучше понять баг.

**Часть 2. Ретест багов.**

**Задачи:**
1. Изучить список багов, которые нашли другие тестировщики в сервисе Mesto.
2. Убедиться, что разработчики исправили баги: провести ретест, используя тестовый стенд.

**Как выполнить:**
1. Создай копию таблицы со списком багов и выполняй задание в ней. https://docs.google.com/spreadsheets/d/1o5CHs6TOsqUlFHU7kRfPS39ZQ1WpbRIsXXJtyJYKkD8/edit?gid=379530441#gid=379530441
2. Если баг исправлен, поставь ему статус «Закрыт». Если баг не исправлен, поставь статус «Переоткрыт». Обрати внимание: возможно, тебе понадобится проставить ещё какие-то статусы, которых нет в выпадающем списке документа. Например, если тестированию мешает другой баг. Добавить новый статус можно так: кликнуть на ячейку «Статус», нажать на карандаш в правом нижнем углу появившегося поля, а затем на кнопку «Создать объект».
3. В комментарии к багам отметь, на каком стенде проведена проверка и устранена ли ошибка. Например: «Проверено на стенде https://code.s3.yandex.net/qa/files/mesto/index.html. Ошибка исправлена». Так ты оставишь резолюцию о проверке.
